```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
```

```python
dataset= pd.read_csv('/content/twitter_training.csv')
dataset.head(10)
```

```python
# Cell 74 contains the following line and will drop 'listed_in(type)'
dataset.drop('2401', axis=1, inplace=True)
```

```python
# Cell 74 contains the following line and will drop 'listed_in(type)'
dataset.drop('Borderlands', axis=1, inplace=True)
```

```python
columns_to_rename = {'im getting on borderlands and i will murder you all ,': 'text'}
```

```python
dataset = dataset.rename(columns=columns_to_rename)
```

```python
dataset.head()
```

```python
columns_to_rename = {'Positive': 'target'}
```

```python
dataset = dataset.rename(columns=columns_to_rename)
```

```python
dataset.head()
```

```python
le=LabelEncoder()
dataset['target']=le.fit_transform(dataset['target'])
```

```python
dataset["text"] = dataset["text"].fillna("")
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

encoded_data = encoder.fit_transform(dataset[['text']])
print(encoded_data)
```

```python
x=dataset['text']
y=dataset['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
vectorizer = TfidfVectorizer()

# Fit the vectorizer to your training data and transform it
x_train_vec = vectorizer.fit_transform(x_train)

# Transform the testing data using the fitted vectorizer
x_test_vec = vectorizer.transform(x_test)

# Now use the vectorized data for training the model
model = SVC()
model.fit(x_train_vec, y_train)
```

```python
model.predict(x_test_vec)
```

```python
print("accuracy score",accuracy_score(y_test,model.predict(x_test_vec)))
```
